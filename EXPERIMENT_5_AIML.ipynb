```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
from google.colab import files
```

```python
uploaded = files.upload()
```

```python
import io
```

```python
df = pd.read_csv(io.BytesIO(uploaded['customers-100.csv']))
```

```python
print("Descriptive Statistics:\n", df.describe())
```

```python
df.hist(figsize=(10, 8), bins=20, color='blue', edgecolor='black')
plt.suptitle('Distribution of Numerical Columns', fontsize=16)
plt.show()
```

The histogram shows the distribution of the values in each numerical column.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.select_dtypes(include=np.number))
plt.title('Boxplot of Numerical Columns', fontsize=16)
plt.show()
```

The boxplot highlights the spread, median, and potential outliers for each numerical column

```python
# Country-wise distribution of customers
country_dist = df['Country'].value_counts()
print(country_dist)
```

```python
# Extracting the year from subscription dates
df['Subscription Date'] = pd.to_datetime(df['Subscription Date']) # Convert 'Subscription Date' column to datetime
df['Subscription Year'] = df['Subscription Date'].dt.year

# Count of customers by subscription year
yearly_customers = df['Subscription Year'].value_counts()
print(yearly_customers)
```

```python
# Plot customer distribution by country
country_dist.plot(kind='bar', figsize=(10, 6), title="Customer Distribution by Country")
plt.show()
```
